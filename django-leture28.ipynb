```python
image_url ='https://static.pexels.com/photos/36753/flower-purple-lical-blosso.jpg'
```

```python
import requests

image_binary = requests.get(image_url).content

with open('flower.jpg', 'wb') as f:
    f.write(image_binary)
```

```python
from PIL import Image #Pillow

flower_image = Image.open('flower.jpg')
```

```python
flower_image
```

```python
from pilkit.processors import Thumbnail

processor = Thumbnail(width =300, height=300)
```

```python
tumb_image = processor.process(flower_image)
```

```python
tumb_image
```

```python
tumb_image.save("tumb-300x300.png")


for quality in [100, 80, 60, 40, 20]:
    tumb_image.save("tumb-300x300-{}.jpg" .format(quality) , quality=quality)
```
